```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot (x,y,color='g', linestyle=':', linewidth=10 )
plt.title ('X vs Y')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show
```

```python
x=np.arange(1,12)
x
```

```python
y1 =2*x
y2 =3*x
```

```python
plt.plot(x,y1,color='r', linestyle=':')
plt.plot(x,y2,color='g',linewidth=5)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,color='r', linestyle=':')

plt.subplot(2,1,2)
plt.plot(x,y2,color='g',linewidth=5)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```
